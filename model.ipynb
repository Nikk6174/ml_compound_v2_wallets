```python
# Cell 5: Advanced feature engineering - create wallet-level features
def create_advanced_wallet_features(df):
    """Create comprehensive wallet-level features for risk assessment"""
    
    wallet_features = []
    
    # Use wallet_address if available, otherwise use 'from' addresses
    if 'wallet_address' in df.columns:
        all_wallets = df['wallet_address'].dropna().unique()
        wallet_col = 'wallet_address'
    else:
        all_wallets = set(df['from'].dropna().unique()) | set(df['to'].dropna().unique())
        wallet_col = None
    
    print(f"Processing features for {len(all_wallets)} wallets...")
    
    for i, wallet in enumerate(all_wallets):
        if i % 1000 == 0:
            print(f"Processed {i} wallets...")
            
        # Get transactions for this wallet
        if wallet_col:
            wallet_txns = df[df[wallet_col] == wallet]
            sent_txns = wallet_txns  # all transactions are from this wallet
            received_txns = df[df['to'] == wallet] if 'to' in df.columns else pd.DataFrame()
        else:
            sent_txns = df[df['from'] == wallet]
            received_txns = df[df['to'] == wallet] 
            wallet_txns = df[(df['from'] == wallet) | (df['to'] == wallet)]
        
        if len(wallet_txns) == 0:
            continue
            
        features = {'wallet_id': wallet}
        
        # === BASIC TRANSACTION FEATURES ===
        features['total_transactions'] = len(wallet_txns)
        features['sent_transactions'] = len(sent_txns)
        features['received_transactions'] = len(received_txns)
        features['send_receive_ratio'] = len(sent_txns) / max(len(received_txns), 1)
        
        # === VALUE FEATURES ===
        features['total_value_sent'] = sent_txns['value'].sum()
        features['total_value_received'] = received_txns['value'].sum() if len(received_txns) > 0 else 0
        features['avg_transaction_value'] = wallet_txns['value'].mean()
        features['max_transaction_value'] = wallet_txns['value'].max()
        features['min_transaction_value'] = wallet_txns['value'].min()
        features['value_std'] = wallet_txns['value'].std()
        features['value_range'] = features['max_transaction_value'] - features['min_transaction_value']
        
        # Zero value transaction ratio (suspicious for certain activities)
        features['zero_value_ratio'] = (wallet_txns['value'] == 0).mean()
        
        # === GAS FEATURES (Advanced Risk Indicators) ===
        if 'gasUsed' in wallet_txns.columns:
            features['avg_gas_used'] = wallet_txns['gasUsed'].mean()
            features['max_gas_used'] = wallet_txns['gasUsed'].max()
            features['gas_efficiency'] = wallet_txns['gasUsed'].std() / max(wallet_txns['gasUsed'].mean(), 1)
        else:
            features['avg_gas_used'] = features['max_gas_used'] = features['gas_efficiency'] = 0
            
        if 'gasPrice' in wallet_txns.columns:
            features['avg_gas_price'] = wallet_txns['gasPrice'].mean()
            features['gas_price_variance'] = wallet_txns['gasPrice'].std()
        else:
            features['avg_gas_price'] = features['gas_price_variance'] = 0
            
        # Total gas cost
        if all(col in wallet_txns.columns for col in ['gasUsed', 'gasPrice']):
            gas_costs = wallet_txns['gasUsed'] * wallet_txns['gasPrice']
            features['total_gas_cost'] = gas_costs.sum()
            features['avg_gas_cost'] = gas_costs.mean()
        else:
            features['total_gas_cost'] = features['avg_gas_cost'] = 0
        
        # === ERROR AND SUCCESS PATTERNS ===
        if 'isError' in wallet_txns.columns:
            features['error_rate'] = wallet_txns['isError'].mean()
            features['total_errors'] = wallet_txns['isError'].sum()
        else:
            features['error_rate'] = features['total_errors'] = 0
            
        if 'txreceipt_status' in wallet_txns.columns:
            features['success_rate'] = wallet_txns['txreceipt_status'].mean()
        else:
            features['success_rate'] = 1
        
        # === TEMPORAL FEATURES ===
        if len(wallet_txns) > 1:
            time_sorted = wallet_txns.sort_values('timeStamp')
            time_diffs = time_sorted['timeStamp'].diff().dropna()
            
            features['avg_time_between_txns'] = time_diffs.mean().total_seconds() / 3600  # hours
            features['min_time_between_txns'] = time_diffs.min().total_seconds() / 3600
            features['max_time_between_txns'] = time_diffs.max().total_seconds() / 3600
            features['time_variance'] = time_diffs.std().total_seconds() / 3600 if len(time_diffs) > 1 else 0
            
            # Activity span
            activity_span = (wallet_txns['timeStamp'].max() - wallet_txns['timeStamp'].min()).days
            features['activity_span_days'] = max(activity_span, 1)
            features['transaction_frequency'] = len(wallet_txns) / features['activity_span_days']
            
            # Burst activity detection (many transactions in short time)
            features['max_daily_transactions'] = time_sorted.groupby(time_sorted['timeStamp'].dt.date).size().max()
            features['burst_ratio'] = features['max_daily_transactions'] / features['transaction_frequency']
        else:
            features['avg_time_between_txns'] = 0
            features['min_time_between_txns'] = 0
            features['max_time_between_txns'] = 0
            features['time_variance'] = 0
            features['activity_span_days'] = 1
            features['transaction_frequency'] = 1
            features['max_daily_transactions'] = 1
            features['burst_ratio'] = 1
        
        # === COUNTERPARTY ANALYSIS ===
        if 'to' in wallet_txns.columns:
            unique_recipients = wallet_txns['to'].nunique()
            features['unique_recipients'] = unique_recipients
            features['recipient_concentration'] = len(sent_txns) / max(unique_recipients, 1)
        else:
            features['unique_recipients'] = 0
            features['recipient_concentration'] = 0
            
        if 'from' in wallet_txns.columns and len(received_txns) > 0:
            unique_senders = received_txns['from'].nunique()
            features['unique_senders'] = unique_senders
        else:
            features['unique_senders'] = 0
            
        features['total_unique_counterparties'] = features['unique_recipients'] + features['unique_senders']
        
        # === FUNCTION/METHOD ANALYSIS ===
        if 'functionName' in wallet_txns.columns:
            unique_functions = wallet_txns['functionName'].nunique()
            features['unique_functions'] = unique_functions
            features['function_diversity'] = unique_functions / len(wallet_txns)
            
            # Most common function ratio
            most_common_func_count = wallet_txns['functionName'].value_counts().iloc[0]
            features['dominant_function_ratio'] = most_common_func_count / len(wallet_txns)
            
            # Contract interaction complexity
            features['contract_complexity'] = 1 - features['dominant_function_ratio']
        else:
            features['unique_functions'] = 0
            features['function_diversity'] = 0
            features['dominant_function_ratio'] = 1
            features['contract_complexity'] = 0
        
        # === PROTOCOL VERSION ANALYSIS ===
        if 'protocol_version' in wallet_txns.columns:
            unique_protocols = wallet_txns['protocol_version'].nunique()
            features['unique_protocols'] = unique_protocols
            features['protocol_diversity'] = unique_protocols / len(wallet_txns)
        else:
            features['unique_protocols'] = 1
            features['protocol_diversity'] = 0
        
        # === BLOCK-LEVEL FEATURES ===
        if 'blockNumber' in wallet_txns.columns:
            features['block_span'] = wallet_txns['blockNumber'].max() - wallet_txns['blockNumber'].min()
            features['avg_blocks_between_txns'] = features['block_span'] / max(len(wallet_txns) - 1, 1)
        else:
            features['block_span'] = 0
            features['avg_blocks_between_txns'] = 0
        
        wallet_features.append(features)
    
    return pd.DataFrame(wallet_features)

# Create comprehensive wallet features
print("Creating advanced wallet-level features...")
wallet_df = create_advanced_wallet_features(df_clean)
print(f"Created features for {len(wallet_df)} wallets")
print(f"\nTotal features created: {len(wallet_df.columns) - 1}")  # -1 for wallet_id
print("\nFeature categories:")
print("- Basic transaction features: 4")
print("- Value features: 7") 
print("- Gas features: 6")
print("- Error/Success features: 3")
print("- Temporal features: 8")
print("- Counterparty features: 5")
print("- Function/Method features: 4")
print("- Protocol features: 2")
print("- Block features: 2")
```

```python
def calculate_advanced_risk_score(df):
    """Calculate sophisticated risk score using multiple risk indicators"""
    
    # Select features for risk calculation (exclude wallet_id)
    feature_cols = [col for col in df.columns if col != 'wallet_id']
    features = df[feature_cols].copy()
    
    # Handle infinite values
    features = features.replace([np.inf, -np.inf], 0)
    
    # Normalize features to 0-1 range
    scaler = MinMaxScaler()
    features_scaled = pd.DataFrame(
        scaler.fit_transform(features), 
        columns=feature_cols, 
        index=features.index
    )
    
    # Define comprehensive risk components with weights
    risk_components = {
        'volume_risk': {
            'features': ['total_transactions', 'total_value_sent', 'max_transaction_value', 'value_range'],
            'weight': 0.20,
            'description': 'High volume activity risk'
        },
        'behavioral_risk': {
            'features': ['send_receive_ratio', 'recipient_concentration', 'transaction_frequency', 'burst_ratio'],
            'weight': 0.25,
            'description': 'Suspicious behavioral patterns'
        },
        'technical_risk': {
            'features': ['avg_gas_used', 'total_gas_cost', 'error_rate', 'gas_efficiency'],
            'weight': 0.20,
            'description': 'Technical complexity and errors'
        },
        'temporal_risk': {
            'features': ['time_variance', 'max_daily_transactions', 'min_time_between_txns'],
            'weight': 0.15,
            'description': 'Unusual timing patterns'
        },
        'diversity_risk': {
            'features': ['total_unique_counterparties', 'function_diversity', 'protocol_diversity', 'contract_complexity'],
            'weight': 0.20,
            'description': 'Interaction diversity patterns'
        }
    }
    
    # Calculate component scores
    risk_scores = pd.DataFrame(index=df.index)
    component_details = {}
    
    for component, config in risk_components.items():
        # Get available features for this component
        available_features = [f for f in config['features'] if f in features_scaled.columns]
        
        if available_features:
            # Calculate component score as weighted average
            component_score = features_scaled[available_features].mean(axis=1)
            risk_scores[component] = component_score * config['weight']
            component_details[component] = {
                'available_features': available_features,
                'weight': config['weight'],
                'description': config['description']
            }
        else:
            risk_scores[component] = 0
            component_details[component] = {
                'available_features': [],
                'weight': 0,
                'description': config['description']
            }
    
    # Calculate base risk score (0-1 scale)
    base_score = risk_scores.sum(axis=1)
    
    # Apply risk adjustments based on specific patterns
    adjusted_score = base_score.copy()
    
    # High error rate penalty
    if 'error_rate' in features_scaled.columns:
        high_error_mask = features_scaled['error_rate'] > 0.1  # >10% error rate
        adjusted_score[high_error_mask] *= 1.3
    
    # Zero value transaction penalty (potential spam or testing)
    if 'zero_value_ratio' in features_scaled.columns:
        high_zero_value_mask = features_scaled['zero_value_ratio'] > 0.5  # >50% zero value txns
        adjusted_score[high_zero_value_mask] *= 1.2
    
    # Very high frequency penalty (potential bot activity)
    if 'transaction_frequency' in features_scaled.columns:
        high_freq_mask = features_scaled['transaction_frequency'] > 0.95  # top 5%
        adjusted_score[high_freq_mask] *= 1.4
    
    # Scale to 0-1000 range
    final_score = np.minimum(adjusted_score * 1000, 1000)
    
    return final_score, risk_scores, component_details

# Calculate advanced risk scores
print("Calculating advanced risk scores...")
risk_scores_final, component_scores, component_info = calculate_advanced_risk_score(wallet_df)

# Add scores to wallet dataframe
wallet_df['base_risk_score'] = risk_scores_final
wallet_df['base_risk_score'] = wallet_df['base_risk_score'].round(0).astype(int)

print("Base risk score distribution:")
print(wallet_df['base_risk_score'].describe())

print("\nRisk component information:")
for component, info in component_info.items():
    print(f"{component}: {info['description']}")
    print(f"  Weight: {info['weight']:.2f}, Features: {len(info['available_features'])}")
```

```python

print("Applying anomaly detection...")

# Prepare features for anomaly detection
feature_cols_anomaly = [col for col in wallet_df.columns if col not in ['wallet_id', 'base_risk_score']]
X = wallet_df[feature_cols_anomaly].fillna(0)

# Handle infinite values
X = X.replace([np.inf, -np.inf], 0)

# Standardize features for anomaly detection
scaler_anomaly = StandardScaler()
X_scaled = scaler_anomaly.fit_transform(X)

# Apply Isolation Forest for anomaly detection
contamination_rate = 0.05  # Expect 5% anomalies
isolation_forest = IsolationForest(contamination=contamination_rate, random_state=42, n_estimators=100)
anomaly_labels = isolation_forest.fit_predict(X_scaled)
anomaly_scores = isolation_forest.score_samples(X_scaled)

# Add anomaly information to dataframe
wallet_df['anomaly_score'] = anomaly_scores
wallet_df['is_anomaly'] = (anomaly_labels == -1)

# Apply K-means clustering for additional pattern detection
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_scaled)
wallet_df['cluster'] = cluster_labels

# Calculate cluster-based risk adjustment
cluster_risk_means = wallet_df.groupby('cluster')['base_risk_score'].mean()
wallet_df['cluster_risk_adjustment'] = wallet_df['cluster'].map(cluster_risk_means) / wallet_df['base_risk_score'].mean()

print(f"Anomalous wallets identified: {wallet_df['is_anomaly'].sum()} ({wallet_df['is_anomaly'].mean()*100:.1f}%)")
print(f"Wallets clustered into {n_clusters} groups")
print("\nCluster risk distribution:")
print(cluster_risk_means.sort_values(ascending=False))
```

```python
def calculate_final_risk_score(df):
    """Calculate final risk score with all adjustments"""
    
    final_scores = df['base_risk_score'].copy()
    
    # Anomaly boost (increase risk for anomalous wallets)
    anomaly_boost = 1.5
    final_scores[df['is_anomaly']] *= anomaly_boost
    
    # Cluster-based adjustment
    final_scores *= df['cluster_risk_adjustment']
    
    # Apply bounds and ensure integer values
    final_scores = np.minimum(np.maximum(final_scores, 0), 1000)
    
    return final_scores.round(0).astype(int)

# Calculate final scores
wallet_df['final_risk_score'] = calculate_final_risk_score(wallet_df)

print("Final risk score distribution:")
print(wallet_df['final_risk_score'].describe())

# Risk level categorization
def categorize_risk(score):
    if score < 200:
        return 'Very Low'
    elif score < 400:
        return 'Low' 
    elif score < 600:
        return 'Medium'
    elif score < 800:
        return 'High'
    else:
        return 'Very High'

wallet_df['risk_category'] = wallet_df['final_risk_score'].apply(categorize_risk)

print("\nRisk category distribution:")
print(wallet_df['risk_category'].value_counts())
```
